```python
import numpy as np
import pandas as pd
import pickle
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn import model_selection
```

```python
data = pd.read_csv("house_price_dataset_new.csv")
data.head()
```

```python
data = data.drop(['society', 'yr_built', 'furniture', 'sale_type', 'amenities', 'market', 'office', 'school', 'college', 'hospital', 'population', 'railway', 'airport', 'on_road', 'air_quality', 'restaurant', 'park'], axis=1)
data.head()
```

```python
data.isnull().sum()
```

```python
df = data.iloc[ : , :-1].values
df
```

```python
label_encoder = LabelEncoder()
```

```python
df[ : ,0] = label_encoder.fit_transform(df[ : ,0])
df
```

```python
mapping = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))
mapping
```

```python
df[ : ,1] = label_encoder.fit_transform(df[ : ,1])
df
```

```python
mapping = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))
mapping
```

```python
df[ : ,2] = label_encoder.fit_transform(df[ : ,2])
df
```

```python
mapping = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))
mapping
```

```python
y = data.price.values
X = df
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y,test_size=0.2)
```

```python
standard_x = StandardScaler()
```

```python
X_train = standard_x.fit_transform(X_train)
X_val = standard_x.transform(X_val)
```

```python
print('Train_Shape: ',X_train.shape)
print("\nX_train:")
X_train
```

```python
print('Val_Shape: ',X_val.shape)
print("\nX_val:")
X_val
```

# RANDOM FOREST REGRESSOR

```python
regressor = RandomForestRegressor(n_estimators = 10, random_state = 0)
```

```python
regressor.fit(X_train, y_train)
```

```python
filename = 'house_model.pkl'
```

```python
pickle.dump(regressor, open(filename, 'wb'))
```

```python
loaded_model = pickle.load(open(filename, 'rb'))
```

```python
accuracy = loaded_model.score(X_val, y_val)
```

```python
print(accuracy*100, '%')
```

```python
predictions = loaded_model.predict(X_val)
```

```python
predictions
```

# LSTM MODEL

```python

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform( X_val)

X_lstm = X_scaled.reshape((X_scaled.shape[0], X_scaled.shape[1], 1))
X_test_lstm = X_test_scaled.reshape((X_test_scaled.shape[0], X_test_scaled.shape[1], 1))

# Build LSTM model
lstm_model = Sequential()
lstm_model.add(LSTM(64, return_sequences=True, input_shape=(X_lstm.shape[1], 1)))
lstm_model.add(LSTM(32))
lstm_model.add(Dense(1))
lstm_model.compile(optimizer='adam', loss='mse')

# Train LSTM model
lstm_model.fit(X_lstm, y_train, epochs=10, batch_size=32, validation_split=0.2)


```

```python
from sklearn.metrics import r2_score, mean_absolute_error

y_pred_lstm = lstm_model.predict(X_test_lstm)
print('LSTM Model Performance:')
print('R2 Score:', r2_score( y_val , y_pred_lstm))
print('Mean Absolute Error:', mean_absolute_error( y_val , y_pred_lstm))
```

# XGBOOST REGRESSOR

```python

from xgboost import XGBRegressor
xgb = XGBRegressor()
xgb.fit(X_train, y_train)

```

```python

y_pred_xgb = xgb.predict(X_val)
print('XGBoost Model Performance:')
print('R2 Score:', r2_score(y_val, y_pred_xgb))
print('Mean Absolute Error:', mean_absolute_error(y_val, y_pred_xgb))
```

```python

from sklearn.ensemble import GradientBoostingRegressor
gb = GradientBoostingRegressor()
gb.fit(X_train, y_train)

```

```python

y_pred_gb = gb.predict(X_val)
print('Gradient Boosting Model Performance:')
print('R2 Score:', r2_score(y_val, y_pred_gb))
print('Mean Absolute Error:', mean_absolute_error(y_val, y_pred_gb))
```

```python
import matplotlib.pyplot as plt
y_pred_lstm = lstm_model.predict(X_test_lstm).flatten()
y_pred_xgb = xgb.predict(X_val)
y_pred_gb = gb.predict(X_val)

plt.figure(figsize=(12, 8))
plt.plot(y_val, label='Actual Values', marker='o')
plt.plot(y_pred_lstm, label='LSTM Predictions', marker='x')
plt.plot(y_pred_xgb, label='XGBoost Predictions', marker='s')
plt.plot(y_pred_gb, label='Gradient Boosting Predictions', marker='d')

plt.title('Model Predictions vs Actual Values')
plt.xlabel('Sample Index')
plt.ylabel('House Price')
plt.legend()
plt.grid()
plt.show()
```
